# Pig

Write a program that allows you to play the following game:


- Each player rolls the dice repeatedly until either a one is rolled, or the player decides to hold.
- If a player rolls a 1, their turn is over; they get no points for the round
- If some other number is rolled, the score is added to their turn total and they are allowed to roll again.
- If the player chooses to hold, they are allowed to keep their turn total; the turn total is added to their overall total score.
- The game is over when one player has a overall total score of 100 or more.

```python
import random
```

## IPO:

Inputs:
- Hold, roll

Process:
- Identify which player is up
- Roll the dice
    - show dice roll
    - ID the value of the roll
    - if its a 1, reset turn total to 0 and end turn
    - if its not a 1, add to turn total
    - add turn total to overall total
- End game when someone's total  is >= 100

Output:
- Result of the roll
- Option to hold or roll again
- Turn total
- Overall total
- Whose turn it is to play

```python
def print_menu():
    print("\n************************")
    print("Welcome to Pig!")
    print("Follow the instructions and choose how you would like to play!")
    print("************************\n")
    return

def roll_dice():
    dice_result = random.randint(1,6)
    return dice_result
        
        
def pig():
    print_menu()
    player_condition = input("Type start to begin\nType stop to exit the game.")
    if player_condition.upper() == "START":
        print("Begin Game")
        Pone_turn = 0
        Ptwo_turn = 0
        Pone_total = 0
        Ptwo_total = 0
        Pturn = 1 
        while Pone_total < 100 and Ptwo_total < 100:
            if Pturn == 1:
                print("Player 1 turn")
                turn_status = input("Type roll to roll or hold to hold and pass turn.")
                if turn_status.upper() == 'ROLL':
                    Pone_roll = roll_dice()
                    if Pone_roll == 1:
                        print("You rolled a 1. Your turn is over.\n")
                        Pone_turn = 0
                        Pturn = 2
                    else:
                        Pone_turn += Pone_roll
                        print("You rolled a", Pone_roll, ", your turn score is now", Pone_turn,".")
                elif turn_status.upper() == 'HOLD':
                    Pone_total += Pone_turn
                    print("You have chosen to hold, your current score is", Pone_total,".\n")
                    Pone_turn = 0
                    Pturn = 2
                else:
                    print("Invalid input, please try again.\n")
            elif Pturn == 2:
                print("Player 2 turn")
                turn_status = input("Type roll to roll or hold to hold and pass turn.")
                if turn_status.upper() == 'ROLL':
                    Ptwo_roll = roll_dice()
                    if Ptwo_roll == 1:
                        print("You rolled a 1. Your turn is over.\n")
                        Ptwo_turn = 0
                        Pturn = 1
                    else:
                        Ptwo_turn += Ptwo_roll
                        print("You rolled a", Ptwo_roll, ", your turn score is now", Ptwo_turn,".")
                elif turn_status.upper() == 'HOLD':
                    Ptwo_total += Ptwo_turn
                    print("You have chosen to hold, your current score is", Ptwo_total,".\n")
                    Ptwo_turn = 0
                    Pturn = 1
                else:
                    print("Invalid input, please try again.\n")
        if Pone_total >= 100:
            print("\nCongrats Player 1, you won!\n The final score was:\n Player 1:", Pone_total, "\n Player 2:", Ptwo_total)
        elif Ptwo_total >= 100:
            print("\nCongrats Player 2, you won!\n The final score was:\n Player 1:", Pone_total, "\n Player 2:", Ptwo_total)
    elif player_condition.upper() == "STOP":
        print("\nLeaving game, thanks for playing!")
    else:
        print("\nInvalid input, please start again")
    
        
```

# Other Games

Try them all with real dice.  Then choose **one** of the following games to simulate:

## 1) Beat It

- This game is played with two players and two dice.
- Each time the dice are rolled, the numbers on the dice are concatenated such that the larger roll becomes the 10’s place and the smaller roll becomes the one’s place. 
    - For instance, a roll of a `3` and a `6` would become `63`
    - In the event that both dice are the same, the same digit is recorded for the 10’s place and the 1’s place.
- In each round, players alternate rolling the dice, as long as they have a higher number.
- The player that has the highest roll for the round gets their points added to their score.
- In the event of a tie, neither player gets any points for that round.
- The game ends when one of the players reaches 100 points.

## 2) Keep the Lowest

- This game is played with two players and five dice.
- Players take turns rolling and selecting dice.
- The dice are rolled and the player *keeps* the lowest, removing one dice from play.
- The player's score is the sum of all their *kept* dice.
- The first player to reach a score of 20 **loses**.

## 3) High/Low

- This game is played with one players and two dice.
- The player starts with \$100 in their account.
- Each round, the user places a wager and a bet
    - The wager is at least \$1, but not *more* than their account balance.
    - The only allowable bets are: 
        - High  (above 7)
        - Low   (below 7)
        - Seven (exactly 7)
- The player rolls a pair of dice.
- A roll is Low if the total of the dice is 2,3,4,5,6;
- A roll is High if the total is 8,9,10,11,12
- Matching Low or High pays out 2:1
- Matching a 7 pays out 4:1
- No match results in the bet being deducted from the account.
- Display the current account balance at the end of each round.
- The game ends when the user has gone bust (no money) or has earned \$250.

## 4) Roll the Bones

- This game is played with one player and two dice.
- The player enters a guess between 5 and 30.
- The player rolls a set of five dice and accumulates the total.
- When the guess is less than the total, award `guess-5` points to the player (this rewards risk taking).
- When the guess is too high, penalize the player `5-guess` points. 
- For example:
    - If the player guess `13` and the roll is `25`, the player wins `8 points. 
    - If the guess was `18` and the roll is `17`, the user earns `-13` points.
- When you guess the total exactly, the player is awarded 50 points.
- Keep track of the total number of points and the average points per turn.
- The game continues until the user wishes to stop.
