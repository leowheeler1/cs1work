```python
import matplotlib.pyplot as plt
```

```python
open_file = open('untitled.txt')
fish = open_file.readlines()

fish_dict = {}

for line in fish[1:]:
    line = line.split(',')
    fish_dict[line[0]] = int(line[1].strip())

datax = fish_dict.keys()
datay = fish_dict.values()

x_axis_locations = range(len(datax))
plt.figure(dpi = 200, figsize = (10,6))
plt.bar(x_axis_locations, datay, align='center', alpha=0.5, color = 'black')
plt.ylabel("Length (cm)")
plt.xlabel("Fish")
plt.xticks(x_axis_locations, datax, rotation='vertical')
plt.show()
```

```python
def sum_maker():
    num_list = []
    big_sum = 0
    average = 0
    while True:
        number = input("Please enter a number to add to the sum, or type \'exit\' to exit :\n")
        
        if number.upper() != 'EXIT':
            try:
                number = int(number)
            except ValueError:
                print('That is not a valid number, please try again')
                continue
            num_list.append(number)
            big_sum += number
            average = big_sum/len(num_list)
            print(f"Your current sum is {big_sum} and the running average is {average}\n")
        else:
            print(f"Your final sum was {big_sum} and the average of your numbers was {average}")
            print(f"Here are all the numbers you used: {num_list}")
            break
    return
```
