# IPO - word list

Input:

- none

Process:

- opens scrabble words
- iterates and adds each word to a set while stripping new lines

Output:

- word list in a set

```python
def create_wordset():
    
    # Opens the file containing the scrabble words
    open_file = open('/srv/data/ScrabbleWords.txt')
    
    # Turns the file into a list that can be iterated over
    words = open_file.readlines()
    
    # Makes the set, and adds each word from the scrabble list into the set, removing the new line indicator (\n)
    wordset = {i.strip('\n') for i in words}
    
    return wordset # Allows the output of the function to be the wordset
```

```python
words = create_wordset() # Creates a variable that will be used in the next function as the wordset
```

```python
words # Shows the set of the words
```

# IPO - spellcheck

Input:

- sentence as a string
- wordset

Process:

- iterates through the sentence and checks words for spelling errors

Output:

- number of misspelled words
- which words are misspelled (correct=lowercase, incorrrect=uppercase)

```python
def spellcheck(sentence, wordset):
    
    # Breaks the sentence into a list so that it can stay ordered and be analyzed
    sentence = sentence.strip('.!?').split(' ')
    
    # Creates blank string for the output sentence
    out_sent = ''
    
    # Counter of spelling errors
    misspell = 0
    
    # Iterates over the sentence
    for i in sentence: # The following three conditionals add the word to the output sentence
        
        if i.upper() in ["I","A"]:        # Compensates for the fact that the scrabble list doesn't have I or A
            out_sent += i.lower() + ' '
        elif i not in wordset:            # Counts up if there is a spelling error, capitalizes the word
            out_sent += i.upper() + ' '
            misspell += 1
        else:                             # Adds the correctly spelled word (lowercase) to the output sentence
            out_sent += i.lower() + ' '
    
    # Send the output telling how many words were misspelled, with the newly created output sentece that just contains
    # the correct words in lowercase and the incorrect words in uppercase
    print(f"You misspelled {misspell} words.\nYour sentence is:\n\n{out_sent}")
    
    return
```
