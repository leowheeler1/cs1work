```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio as audio
from ipywidgets import widgets
```

```python
sampling_rate = 32000
duration = 2
note_length = int(sampling_rate*duration)
t = np.linspace(0, duration, note_length)
```

```python
x = np.sin(440 * (2*np.pi*t))
```

```python
plt.figure(dpi=100, figsize=(12,4))
x = np.sin(440 * (2*np.pi*t))
y = np.sin(554 * (2*np.pi*t))
z = np.sin(659 * (2*np.pi*t))
c = [x[i]+y[i]+z[i] for i in range(len(x))]
plt.plot(t[0:200],x[0:200], linewidth=4)
plt.plot(t[0:200],y[0:200], linewidth=4)
plt.plot(t[0:200],z[0:200], linewidth=4)
plt.show()
```

```python
audio(x, rate=sampling_rate, autoplay=True)
```

```python
audio(y, rate=sampling_rate, autoplay=True)
```

```python
audio(z, rate=sampling_rate, autoplay=True)
```

```python
def piano_key_freq(N):
    note = 440 * 2**((N-49)/12)
    return note
```

```python
audio(c, rate=sampling_rate, autoplay=True)
```

```python
def synth(note):
    x = np.sin(note*(2*np.pi*t))
    display(audio(x, rate = sampling_rate, autoplay = True))
```

```python
synth(220)
```

```python
song1 = [56,54,52,54,
        56,56,56,56,
        54,54,54,54,
        56,59,59,59,
        56,54,52,54,
        56,56,56,56,
        54,54,56,54,
        52,52,52,52]
```

```python
duration = 0.5
note_length = int(sampling_rate * duration)
t = np.linspace(0, duration, note_length)
```

```python
play = []
for i in song1:
    x = np.sin(piano_key_freq(i)*(2*np.pi*t))
    play = play + list(x)
```

```python
audio(play, rate=sampling_rate, autoplay=True)
```
