## Before we start!

- There are probably hundreds of different plotting libraries
- We will focus on just one for now
- We will be duscyssing the library called MatPlotLib
- Almost definitely any plot that can be constructed can be done so in matplotlib
- Only does static plots

```python
import matplotlib.pyplot as plt
```

- The "as plt" is optional
- It is a very common convention

```python
plt.plot([1,2,3,4,5])
plt.show()
```

- When plt.plot is called with a single list, it assumes that list is Y values
- The x values are assumed to be the list indices
- Straight lines are drawn between the plotted points

```python
plt.plot([1,2,3,4,5], [10,5,10,5,10])
plt.title("This is the Title!")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.grid()
plt.show()
```

- When the plot command is given two lists, it assumes list 1 is X values, and 2 is Y values
- The first list and second list must have the same number of elements

```python
for i in range(5):
    print(i, 
          "These are some words.")
```

```python
plt.plot([1, 2, 3,4, 5, 4, 3, 2, 1], 
         [10,5,10,5,10,15,10,15,10],
        color = "goldenrod",
        linewidth = 3,
        linestyle = "-",
        marker = ".",
        markersize = 15,
        markerfacecolor = "steelblue",
        markeredgecolor = "black")
plt.title("This is the Title!")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.xlim(0,6)
plt.ylim(3,17)
plt.grid()
plt.show()
```

```python
x_list = [-5.0]
y_list = [x_list[-1]**2]

for i in range(100):
    last_x = x_list[-1]
    x_list.append(last_x+0.1)
    y_list.append(x_list[-1]**2)
```

```python
plt.plot(x_list,y_list,
        color = "cornflowerblue",
        marker = ".",
        markersize = 4,
        markerfacecolor = "black",
        markeredgecolor = "black")
plt.show()
```
