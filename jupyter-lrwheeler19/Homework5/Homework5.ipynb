# IPO zip code dictionary

Input:

- none

Process:

- open file
- read the file
- establish an empty dictionary
- clean up lines and fill the dictionary
    - strip new lines and seperate each line into a list
    - add the zip code as the key and a tuple with city and state as value
- close file

Output:

- dictionary with zip codes as keys, and a tuple containing city and state as the value

```python
def get_zip_dict(): # <------------------------ Creates dictionary with zip codes as keys and city/state as value
    open_file = open('/srv/data/ZipCodes.csv')
    zip_codes = open_file.readlines() # <------ Seperates the file into a list
    zip_dict = {} # <-------------------------- Establishes blank dictionary
    for line in zip_codes: 
        line = line.strip().split(",") # <----- Takes the '\n' off of the line and then makes each line its own list
        zip_dict[line[0]] = (line[1],line[2]) # Puts into the dictionary (1) the zip code as key and (2) the city and state
    open_file.close()                         # as values
    return zip_dict
```

# IPO

Input:

- none 

Process:

- open file
- read file
- establish blank dictionary
- add values to dictionary
    - strip new lines and split each line into a list
    - add the values to the dictionary
        - full state name is key, abbreviation is dictionary

Output:

- dictionary with full state name as key and abbreviation as value

```python
def get_state_abbrev(): # <----------------- Makes dictionary with full state name as key and abbreviation as value
    open_file = open('/srv/data/abbrev.csv')
    abbrev = open_file.readlines() # <------ Makes the file into a list
    abbrev_dict = {} # <-------------------- Establishes blank dictionary 
    for line in abbrev:
        line = line.strip().split(",") # <-- Same purpose as in `get_zip_dict()`
        abbrev_dict[line[0]] = line[1] # <-- Adds full state name as key and abbreviation as value into the dictionary
    return abbrev_dict
```

# IPO

Input:

- zip code as as string

Process:

- reestablish the zip code dictionary
- determine if the input is a valid zip code
    - If not, output zip code not found
    - Otherwise:
        - create a tuple for the city and state
        - abbreviate the state name
        - assign the city, abbreviated state name, and zip code to a tuple

Output:

- tuple containing city, state abbreviation, and zip code
    - unless zip_code is not found, then output a 'Not Found' statement

```python
def lookup(zip_code): # <----------------------- Takes zip code as input and outputs city, state abbreviation and zip code
    zip_dict = get_zip_dict() # <--------------- Reestablishes the dictionary for zipcodes
    if zip_code not in zip_dict.keys(): # <----- Checks to see if the zip code is valid
        output = zip_code + " Not Found." # <--- Output if zip code is not valid
    else:
        city_state = get_zip_dict()[zip_code] # <--------- Creates a tuple with city and state
        state_abbrev = get_state_abbrev()[city_state[1]] # Gets state abbreviation with the state name from city_state
        output = (city_state[0], state_abbrev, zip_code) # Creates tuple with city, abbreviated state, and zip code
    return output
```
