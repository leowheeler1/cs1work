Create a function which determines the location of a guessed letter in the secret word.  The function should return a *list* of such indices.  If there are no indices, it should return a blank list.

```python
def location_of_letter_in_word(guessed_letter, secret_word):
    # Input: -guessed_letter, the letter that was guessed.  Should be a string.
    #        -secret_word, the secret word we are trying to guess. Should be a string.
    # Process: Loop through the indices of the secret word and if the guessed_letter 
    #           matches the secret word, save that index to a list
    # Output: Return the list of indices that were found
```

```python
def location_of_letter_in_word(guessed_letter, secret_word):
    letter_locations = []
    for i in range(len(secret_word)):
        if guessed_letter.upper() == secret_word.upper()[i]:
            letter_locations.append(i)
    return letter_locations
```

#### Problem 6
By this problem, we should have all the parts needed to make our Hangman game.  Your job is to write a main function, using all your previously made functions, to play a full game of Hangman.

Before you go ahead, reflect on these points:

- You could take many turns.  How should we implement the repeating turns?
- The game stops because:
    - We guessed the full word and have won, in which case there are no more missing letters
    - We ran out of wrong guesses (the standard version ends on the 7th wrong guess)

```python
def play_hangman():
    # Input: None
    # Process: -Initialize all the important variables (secret word, guess list, etc.)
    #          -Starts taking turns (in a loop)
    #          -For each turn:
    #             -display the current state of the word
    #             -display the previously geussed letters
    #             -ask the user for a new letter (tell them if it's already been guessed)
    #             -update the guessed word with the new letter
    #             -display the number of incorrect guesses (game ends when it reaches 7)
```

```python
def play_hangman():
    wordlist = word_list()
    secret_word = pick_secret_word(wordlist)
    guessed_word = ["_"]*len(secret_word)
    guessed_letters = [] # Initializing list
    wrong_guesses = 0
    guesses_left = 7
    print(":DEBUG MODE: :DEBUG MODE: :DEBUG MODE:")
    print("Secret Word: ", secret_word) #debugging, take out later
    print(":DEBUG MODE: :DEBUG MODE: :DEBUG MODE:")
    while guesses_left > 0:
        print("\n \n")
        display_guessed_letter_list(guessed_letters)
        valid_guesses = valid_guess_list()
        guess = input("Please guess a letter: ")
        guess = guess.upper()
        
        if guess not in valid_guesses:
            print("That is not a valid guess, please try again")
            continue
            
        elif guess not in guessed_letters:
            letter_locs = location_of_letter_in_word(guess, secret_word)
            guessed_word = update_guessed_word(guessed_word, guess, letter_locs)
            guessed_letters.append(guess)
            
            if guess not in secret_word:
                print(guessed_word)
                print("That letter is not in the word!")
                wrong_guesses += 1
                guesses_left -= 1
                print("You have made", wrong_guesses, "wrong guesses, and have", guesses_left, "wrong guesses left.")
            else:
                print(guessed_word)
                print(guess, "was in the word! You have", guesses_left, "wrong guesses left.")
                if "_" not in guessed_word:
                    break
        elif guess in guessed_letters:
            print("You have guessed that letter already, please guess again.")
            continue
        
        elif "_" not in guessed_word:
            break
    if wrong_guesses < 7:
        print("Congrats!! You have won the game!")
        print("The word was: ", secret_word.lower())
    elif wrong_guesses == 7:
        print("You ran out of guesses, you lose!")
        print("The word was: ", secret_word.lower())
```

```python
play_hangman()
```
