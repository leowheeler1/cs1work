```python
def generate_citation(ID, speed, speedLimit, construction):
    print(str(ID), end = "")
    speed -= speedLimit
    if speed < 3:
        print(" Compliant")
    elif 3 <= speed < 8:
        print(" Warning")
    elif speed >= 8:
        Fine = min((50 + speed*5), 200)*(1 + int(construction))
        print(" Fine $" + str(Fine))
```

```python
generate_citation("ILF-318", 35, 25, False)
```
